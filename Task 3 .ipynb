### Feature Selection for Analysis :

```python
# Alternatively, we can proceed as follows :
Age_Income_data = [data['age'],data['income']]
Age_Income_data = pd.DataFrame(Age_Income_data)
```

```python
Age_Income_data.head()
```

```python
Age_Income_data = Age_Income_data.transpose()
Age_Income_data.head()
```

# Data Visualization :

### Bar Chart Visualization of Age vs Income :

```python
plt.bar(data['age'],data['income'])
plt.title('Bar Plot Representation of Age-Income Statistics')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

```python
# Alternatively,we can proceed as follows :
Age_Income_data.plot(kind='bar',figsize=(10,6))
plt.title('Bar Plot Representation of Age-Income Statistics')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

### Line Chart Representation of Age Vs Income :

```python
plt.plot(data['age'])
plt.plot(data['income'])
plt.title("Line Chart Representation of Age vs Income")
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

```python
# Alternatively, we can proceed as follows :
Age_Income_data.plot(kind='line',figsize=(10,6))
plt.title('Line Chart Representation of Age-Income Statistics')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

### Scatter Plot Representation of Age vs Income :

```python
plt.scatter(data['age'],data['income'])
plt.title("Scatter Plot Representation of Age-Income Statistics")
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

```python
# Alternatively, we can proceed as follows :
Age_Income_data.plot(kind='scatter',x='age',y='income',figsize=(9,5))
plt.title('Scatter Plot Representation of Age-Income Statistics')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

### Histogram Representation of Income Statistics :

```python
plt.hist(data['income'])
plt.title('Histogram of Income Distribution')
plt.ylabel("Frequency")
plt.xlabel('Income Profile')
plt.show()
```

```python
plt.figure(figsize=(10,6))
Age_Income_data['income'].plot(kind='hist')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```
