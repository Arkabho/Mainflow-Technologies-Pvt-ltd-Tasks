#### Handling Missing Values :

```python
print(dataset['ChipRate'].mean())
print(dataset['ChipRate'].median())
```

```python
dataset['ChipRate'] = dataset['ChipRate'].fillna(dataset['ChipRate'].mean())
```

```python
print(dataset['BF-CMratio'].mean())
print(dataset['BF-CMratio'].median())
```

```python
dataset['BF-CMratio'] = dataset['BF-CMratio'].fillna(dataset['BF-CMratio'].mean())
```

```python
print(dataset['BlowFlow'].mean())
print(dataset['BlowFlow'].median())
```

```python
dataset['BlowFlow'] = dataset['BlowFlow'].fillna(dataset['BlowFlow'].median())
```

```python
print(dataset['UCZAA'].mean())
print(dataset['UCZAA'].median())
```

```python
dataset['UCZAA'] = dataset['UCZAA'].fillna(dataset['UCZAA'].median())
```

```python
print(dataset.isna().sum())
```

```python
dataset.columns = dataset.columns.str.strip()
```

```python
print(dataset.columns)
```

```python
print(dataset['AAWhiteSt-4'].mean())
print(dataset['AAWhiteSt-4'].median())
```

```python
dataset['AAWhiteSt-4'] = dataset['AAWhiteSt-4'].fillna(dataset['AAWhiteSt-4'].mean())
```

```python
print(dataset['SulphidityL-4'].mean())
print(dataset['SulphidityL-4'].median())
```

```python
dataset['SulphidityL-4'] = dataset['SulphidityL-4'].fillna(dataset['SulphidityL-4'].mean())
```

```python
# To check for existence of Missing Values in the given dataset again :-
```

```python
print(dataset.isnull().sum())
```

```python
dataset.dropna(inplace=True)
dataset.reset_index(drop=True,inplace=True)
```

```python
print(dataset.isna().sum())
```

```python
# Therefore, now we have a completely curated dataset that does not contain any missing values
```

```python
print(dataset.shape)
```

```python
# Therefore, the given dataset contains 318 rows and 23 columns
```
